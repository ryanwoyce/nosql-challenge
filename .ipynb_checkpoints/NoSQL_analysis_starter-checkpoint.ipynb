### 1. Which establishments have a hygiene score equal to 20?

```python
establishments.find_one()
```

```python
# Find the establishments with a hygiene score of 20
pipeline = [{'$match': {'scores.Hygiene': 20}}]
result = list(establishments.aggregate(pipeline))
# Use count_documents to display the number of documents in the result
print(f'Number of documents: {len(result)}')
print('--------------------------------------')
# Display the first document in the results using pprint
print('First document:')
print('')
pprint(result[0:1])
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd
hy20_df = pd.DataFrame(result)
# Display the number of rows in the DataFrame
print(f'Number of rows in DataFrame: {len(hy20_df)}')
# Display the first 10 rows of the DataFrame
hy20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = [{'$match': {'RatingValue': {'$gte': 4}}}]
results = list(establishments.aggregate(query))
# Use count_documents to display the number of documents in the result
print(f'Number of documents: {len(results)}')
print('--------------------------------------')
# Display the first document in the results using pprint
print('First document:')
print('')
pprint(results[0:1])
```

```python
# Convert the result to a Pandas DataFrame
rating_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f'Number of rows: {len(rating_df)}')
# Display the first 10 rows of the DataFrame
rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {'RatingValue': 5, 
         'geocode.latitude': {'$lte': (latitude + degree_search), '$gte': (latitude - degree_search)}, 
         'geocode.longitude': {'$lte': (longitude + degree_search), '$gte': (longitude - degree_search)} 
         }
sort = [('scores.Hygiene', 1)]


# Print the results
for result in establishments.find(query).sort(sort).limit(5):
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
final_df = pd.DataFrame(establishments.find(query).sort(sort).limit(5))
final_df
```
